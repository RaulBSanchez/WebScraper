```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url = 'https://pitchfork.com/best/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
best_albums_music = soup.find("div", id="best-new-albums")
best_new_tracks = soup.find("div", id = "best-new-tracks")
best_new_music = soup.find("div", class_="SubTopicDiscoveryWrapper-iRTUpJ kBTops" )
```

```python
def extract_best_new_music(best_new_music):
    new_music = []
    new_music_items = best_new_music.find_all("div", class_="SubtopicDiscoveryHedContainer-ehGPZM")

    #print(new_music_items)
    for item in new_music_items:


        category_tag = item.find("span", class_= "RubricName-fVtemz")
        category = category_tag.get_text(strip = True) if category_tag else "Unknown Category"
        
        artist_tag = item.find("div", class_="SummaryItemSubHedBase-gMyBBg")
        artist = artist_tag.get_text(strip=True) if artist_tag else "Unknown Artist"
        
        title_tag = item.find("h3", class_="SummaryItemHedBase-hiFYpQ")
        title = title_tag.get_text(strip=True) if title_tag else "Unknown Title"

        genre_tag = item.find("span", class_= "RubricName-fVtemz")
        genre = genre_tag.get_text(strip=True) if genre_tag else "Unknown Genre"

        date_tag = item.find("time")
        publish_date = date_tag.get_text(strip=True) if date_tag else "Unknown Date"
    

        new_music.append([category, artist, title, genre, publish_date])

# Create DataFrame
    df = pd.DataFrame(new_music, columns=["Category", "Artist", "Album",  "Genre", "Publish Date"])
    return df

extract_best_new_music(best_new_music)
```

```python
def extract_best_albums(best_albums_music):
    albums = []
    albums_items = best_albums_music.find_all("div", class_="SummaryItemWrapper-iwvBff")
    #print(reviews)
    #print(albums_items[0])
    
    for item in albums_items:

        artist_tag = item.find("div", class_="SummaryItemSubHedBase-gMyBBg")
        artist = artist_tag.get_text(strip=True) if artist_tag else "Unknown Artist"
        
        title_tag = item.find("h3", class_="SummaryItemHedBase-hiFYpQ")
        title = title_tag.get_text(strip=True) if title_tag else "Unknown Title"

        genre_tag = item.find("span", class_= "RubricName-fVtemz")
        genre = genre_tag.get_text(strip=True) if genre_tag else "Unknown Genre"
    

        desc_tag = item.find("div", class_="SummaryItemDek-CRfsi")
        description = desc_tag.get_text(strip=True) if desc_tag else "No Description"

        date_tag = item.find("time")
        publish_date = date_tag.get_text(strip=True) if date_tag else "Unknown Date"

        albums.append([artist, title, genre,publish_date])

# Create DataFrame
    df = pd.DataFrame(albums, columns=["Artist", "Album",  "Genre", "Publish Date"])
    return df
    
```

```python
extract_best_albums(best_albums_music)
```

```python
def extract_best_tracks(best_new_tracks):
    tracks = []
    track_items = best_new_tracks.find_all("div", class_="SummaryItemContent-eiDYMl")
    #print(reviews)
    #print(albums_items[0])
    
    for track in track_items:
        

        artist_tag = track.find("div", class_="SummaryItemSubHedBase-gMyBBg")
        artist = artist_tag.get_text(strip=True) if artist_tag else "Unknown Artist"

        #track name
        song_tag = track.find("h3", class_="SummaryItemHedBase-hiFYpQ")
        #print(track_tag)
        song = song_tag.get_text(strip=True) if song_tag else "Unknown Track"

        #Genre
        genre_tag = track.find("span", class_="RubricName-fVtemz")
        genre = genre_tag.get_text(strip=True) if genre_tag else "Unknown Genre"

        #descritipon


        date_tag = track.find("time")
        publish_date = date_tag.get_text(strip=True) if date_tag else "Unknown Date"

        tracks.append([artist, song, genre, publish_date, ])

# Create DataFrame
    df = pd.DataFrame(tracks, columns=["Artist", "Track", "Genre", "Publish Date"])
    return df
    
extract_best_tracks(best_new_tracks)
```
